```python
from matplotlib.pyplot import MultipleLocator

fig, ax = plt.subplots(figsize=(20, 20))

plt.title("Evaluation loss comparison", fontsize=30)
plt.xlabel("Type", fontsize=18)
plt.ylabel("Loss", fontsize=18)

x = [1, 2, 3, 4, 5, 6, 7, 8]

label = ['Original', 'Gain', 'TimeStrnch', 'PitchShift', 'Gain + Time', 'Gain + Pitch', 'Time + Pitch' , 'Gain + Time + Pitch']   # 標籤數據
loss = [0.007005349267,	0.01322164387,	0.0004538419016,	0.0002504740551,	0.000081,	0.0000925,	0.00006,	0.0003355058143]
wer = [8.48045977, 9.928735632, 0.7310344828, 0.5103448276, 0.004597701149, 0.1287356322, 0.004597701149, 0.9586206897]
plt.bar(x, loss,width=0.5) 

y_major_locator=MultipleLocator(0.001)
#把y轴的刻度间隔设置为10，并存在变量里
ax.yaxis.set_major_locator(y_major_locator)
#把y轴的主刻度设置为10的倍数
plt.ylim(0, 0.0135)

plt.xticks(ticks=x,
          labels=label,
          fontsize=15,
          rotation=30) 

my_formatted_list = [ '%.6f' % elem for elem in loss ]
A = [float(x) for x in my_formatted_list]

def add_value_label(x_list,y_list):
    for i in range(1, len(x_list)+1):
        plt.text(i,y_list[i-1],y_list[i-1], ha="center", fontsize=12)

add_value_label(x, A)

plt.show()
```
